```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1000, 1000)
beta = np.linspace(0.0001, 0.02, num=1000)
beta_sqrt = np.sqrt(1-beta)
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].plot(x, beta)
axs[0].set_title("beta")
axs[1].plot(x, beta_sqrt)
axs[1].set_title("sqrt(1 - beta)")
```
